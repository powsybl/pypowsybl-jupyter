# Adaptive Text Zoom Demo

This notebook demonstrates how to use the `adaptive_text_zoom` parameter in various widgets to enable or disable the adaptive text zoom feature in Network Area Diagrams (NAD).


## 1. Using `adaptive_text_zoom` in `display_nad`


```python
import pypowsybl as pp
from pypowsybl_jupyter import display_nad

network = pp.network.create_four_substations_node_breaker_network()
diagram = network.get_network_area_diagram()

# Display with adaptive text zoom enabled (default)
display_nad(diagram, adaptive_text_zoom=True)
```

```python
# Display with adaptive text zoom disabled
display_nad(diagram, adaptive_text_zoom=False)
```

## 2. Using `adaptive_text_zoom` in `nad_explorer`


```python
from pypowsybl_jupyter import nad_explorer

# nad_explorer with adaptive text zoom disabled
nad_explorer(network, adaptive_text_zoom=False)
```

## 3. Using `adaptive_text_zoom` in `network_explorer`


```python
from pypowsybl_jupyter import network_explorer

# network_explorer with adaptive text zoom disabled
network_explorer(network, adaptive_text_zoom=False)
```
